```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline

# Assuming 'X_train', 'X_test', 'y_train', 'y_test' are your training and testing data

# Define the columns to be scaled (assuming you have numerical features)
numeric_features = ['Floor', 'Bathroom', 'Bedroom', 'Land', 'Road']

# Create a ColumnTransformer to apply the scaling only to numerical features
column_trans = ColumnTransformer([('num', StandardScaler(), numeric_features)])

# Create the LinearRegression model
lr = LinearRegression()

# Create the pipeline
pipe = make_pipeline(column_trans, lr)

# Fit the pipeline on the training data
pipe.fit(X_train, y_train)

# Make predictions on the test data
y_pred_lr = pipe.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE) and R-squared (R2)
mse = mean_squared_error(y_test, y_pred_lr)
r2 = r2_score(y_test, y_pred_lr)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
print("Root Mean Squared Error:", rmse)



# Perform prediction on new data
new_data = pd.DataFrame({
    'Floor': [2.5],
    'Bathroom': [3.0],
    'Bedroom': [4],
    'Land': [1500.0],
    'Road': [12.0]
})

predicted_price = pipe.predict(new_data)
print("Predicted Price:", predicted_price)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline

# Load and preprocess your data (X_train, X_test, y_train, y_test)

numeric_features = ['Floor', 'Bathroom', 'Bedroom', 'Land', 'Road']
categorical_features = ['Address', 'Face']

column_trans = ColumnTransformer([
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

random_forest = RandomForestRegressor()

pipe = make_pipeline(column_trans, random_forest)

# Define the parameter distributions for RandomizedSearchCV
param_dist = {
    'randomforestregressor__n_estimators': np.arange(100, 1000, 100),
    'randomforestregressor__max_depth': [None, 10, 20, 30],
    'randomforestregressor__min_samples_split': [2, 5, 10],
    'randomforestregressor__min_samples_leaf': [1, 2, 4],
    'randomforestregressor__max_features': ['auto', 'sqrt', 'log2'],
    'randomforestregressor__bootstrap': [True, False]
}

# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(pipe, param_distributions=param_dist, n_iter=100,
                                   cv=5, scoring='neg_mean_squared_error', n_jobs=-1, random_state=42)

# Fit the randomized search on the training data
random_search.fit(X_train, y_train)

# Get the best estimator and best parameters
best_estimator = random_search.best_estimator_
best_params = random_search.best_params_

# Make predictions on the test data using the best estimator
y_pred_best = best_estimator.predict(X_test)

# Calculate evaluation metrics
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)
mae_best = mean_absolute_error(y_test, y_pred_best)

print("Best Random Forest Regression - Mean Squared Error:", mse_best)
print("Best Random Forest Regression - R-squared:", r2_best)
print("Best Random Forest Regression - Mean Absolute Error:", mae_best)

# Perform prediction on new data using the best estimator
new_data_rf = pd.DataFrame({
    'Floor': [2.5],
    'Bathroom': [3.0],
    'Bedroom': [4],
    'Land': [1500.0],
    'Road': [12.0],
    'Address': ['Aakasedhara'],
    'Face': ['South West']
})

predicted_price_rf = best_estimator.predict(new_data_rf)
print("Random Forest Regression - Predicted Price:", predicted_price_rf)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import RandomizedSearchCV, cross_val_predict
# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(pipe, param_distributions=param_dist, n_iter=100,
                                   cv=5, scoring='neg_mean_squared_error', n_jobs=-1, random_state=42)

# Fit the randomized search on the training data
random_search.fit(X_train, y_train)

# Get the best estimator and best parameters
best_estimator = random_search.best_estimator_
best_params = random_search.best_params_

# Print the best hyperparameters
print("Best Hyperparameters:")
for param, value in best_params.items():
    print(f"{param}: {value}")
print("-----------------------------")

# Perform cross-validation on the best estimator
cv_results = cross_val_predict(best_estimator, X_train, y_train, cv=5)

# Calculate evaluation metrics for each fold
mse_cv = mean_squared_error(y_train, cv_results)
r2_cv = r2_score(y_train, cv_results)
mae_cv = mean_absolute_error(y_train, cv_results)

# Print the evaluation metrics for each fold
for fold, (train_idx, val_idx) in enumerate(random_search.cv_results_['split_indices']):
    fold_X_train, fold_X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
    fold_y_train, fold_y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    best_estimator.fit(fold_X_train, fold_y_train)
    fold_pred = best_estimator.predict(fold_X_val)
    
    fold_mse = mean_squared_error(fold_y_val, fold_pred)
    fold_r2 = r2_score(fold_y_val, fold_pred)
    fold_mae = mean_absolute_error(fold_y_val, fold_pred)
    
    print(f"Fold {fold+1} - Mean Squared Error: {fold_mse}")
    print(f"Fold {fold+1} - R-squared: {fold_r2}")
    print(f"Fold {fold+1} - Mean Absolute Error: {fold_mae}")
    print("-----------------------------")

# Print overall cross-validation metrics
print("Overall Cross-Validation Metrics:")
print("Mean Squared Error:", mse_cv)
print("R-squared:", r2_cv)
print("Mean Absolute Error:", mae_cv)
```
